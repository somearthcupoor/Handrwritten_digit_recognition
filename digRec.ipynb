```python
y_count =  y.to_numpy()
a=np.unique(y_count,return_counts=True)
plt.figure(figsize=(10,10))
plt.plot(np.reshape(a[1],(10,1)))
plt.bar(a[0],a[1], color ='orange',
        width = 0.8)
```

```python
from sklearn.model_selection import  train_test_split
#80% train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
list=[3,5,7,9,11]
plt.figure(figsize=(15, 10))

plt.suptitle("KNN Confusion matrix", fontsize=18, y=1)

for count,i in enumerate(list):
    print(i)
    
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(x_train, y_train)

    y_pred=model.predict(x_val)


    print('accuracy = ',accuracy_score(y_val, y_pred),'\n')
    print(classification_report(y_val, y_pred))

    #Confusion Matric
    conf_matrix = confusion_matrix(y_true=y_val, y_pred=y_pred)
    ax = plt.subplot(2,3,count + 1)
    ax.set_title(count+1)
    ax.matshow(conf_matrix,alpha=.5)
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center')
    ax.xaxis.tick_bottom()
```
